```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plt.style.use('seaborn')
```

if we have data (x<sub>1</sub>,y<sub>1</sub>), (x<sub>2</sub>,y<sub>2</sub>), (x<sub>3</sub>,y<sub>3</sub>), ...,  (x<sub>n</sub>,y<sub>n</sub>), and we want to plot then in a scatter plot, we devide the data into two lists, the first list for the x-coorinates [x<sub>1</sub>, x<sub>2</sub>, ..., x<sub>n</sub>] and the second list for the y-coordinates [y<sub>1</sub>, y<sub>2</sub>, ..., y<sub>n</sub>], and then pass these two lists to `plt.scatter()`.

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x,y)
```

To change the size of the dots, use the attribute `s` that takes a numeric value.

```python
plt.scatter(x,y, s=100)
```

To change the shape of points locations from the default shape (dot) to another available options, use the attribute `marker` that takes several options like `x` for example. In addition to that, we can set the color of the marker usong the attribute `c` and set it either to a color string or a RGB value.

```python
plt.scatter(x,y, s=100, c='green', marker='x')
```

we can set a border around dots by setting `edgecolor` attribute to a string color or a RGB color. In addition to contorl the opacity of dots using `alpha` attribute.

```python
plt.scatter(x,y, s=100, c='green', edgecolors='black', linewidths=1, alpha=.75)
```

Instead of setting one color to all the plot dots, we can set shades of a color, each shade to each dot based on a list of shade intensity list that each value in it corresponds to its counterpart dot at the same index from the x and y lists.

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
```

```python
plt.scatter(x,y, s=100, c=colors, edgecolors='black', linewidths=1, alpha=.75)
```

The color of dots when setting `c` attribute to a color intensity list will be a grayscale by default. To set a color for the shades, use the attribute `cmap` that accepts some predefined options like `Greens`.

```python
plt.scatter(x,y, s=100, c=colors, cmap='Greens', edgecolors='black', linewidths=1, alpha=.75)
```

To give indication about the meaning of each shade, set a variable `var_name` to plt.colorbar() and then set `var_name.set_label('explainatory title')`

```python
plt.scatter(x,y, s=100, c=colors, cmap='Greens', edgecolors='black', linewidths=1, alpha=.75)
cbar = plt.colorbar()
cbar.set_label('satisfaction level')
```

we can control the sizes of dots the same way as we controlled the colors of dots, by using `s` attribute.

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
        538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x,y, s=sizes, c=colors, cmap='Greens', edgecolors='black', linewidths=1, alpha=.75)
cbar = plt.colorbar()
cbar.set_label('satisfaction level')
```

```python
data = pd.read_csv('data/scatterData.csv')
```

```python
data.head()
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, edgecolors='black', linewidths=1, alpha=0.75)
plt.title('Trending Youtube video Corey Schafer channel')
plt.xlabel('View counts')
plt.ylabel('Total Likes')
```

```python
plt.scatter(view_count, likes, edgecolors='black', linewidths=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube video Corey Schafer channel')
plt.xlabel('View counts')
plt.ylabel('Total Likes')
```

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolors='black', linewidths=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.title('Trending Youtube video Corey Schafer channel')
plt.xlabel('View counts')
plt.ylabel('Total Likes')
```
