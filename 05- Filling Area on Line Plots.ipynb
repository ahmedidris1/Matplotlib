```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data/dataPlotlip.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries)

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=.25)

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
overall_median = 57287
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, overall_median, alpha=.25)

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=.25)

plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries <= overall_median), 
                 interpolate=True, alpha=.25)

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=.25)

plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries <= overall_median), 
                 interpolate=True, color='red', alpha=.25)

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha=.25, label='Above Avg.')

plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color='red', alpha=.25, label='Below Avg.')

plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```
