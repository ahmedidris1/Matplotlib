```python
from matplotlib import pyplot as plt
```

### Pie Charts      
It is similar to `Bar Charts` the difference is, instead of showing an exact value for each category in x-axis, it shows the ratio of each categorey from the whole Pie chart. In other words, as the name implies (Pie Chart) we can think of the whole chart as a pie of 100% and the each category is a slice of it.

```python
plt.style.use('fivethirtyeight')
```

If we have the following set of ratios `slices` = `[60, 40]`.

```python
slices = [60, 40]
```

We can plot these ratios `slices` in a Pie chart using the method `plt.pie()` and pass to it as an argument the list of ratios `slices`.

```python
plt.pie(slices)
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

As we see, there are no labels to indicate the meaning for each segment of the Pie chart. To label each segment of the Pie chart we can pass an argument, a list of labels, to an attribute of the method `plt.pie()` called `labels`. 

```python
labels = ['sixty', 'forty']
```

```python
plt.pie(slices, labels=labels)
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

Note that, it is not a must that the total sum of `slices` values has to be a value of 100. It can be any number. `matplotlib` will take care of finding the right ratio of each value out of 100%. Therefore, if we have `slices` = `[120, 80]`, it is clear they are the duble of `slices` = `[60, 40]` and thus, the ratio still be same 60% to 40%, resulting in the same chart.

```python
slices = [120, 80]
```

```python
plt.pie(slices, labels=labels)
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

There is an attribute of `plt.pie()` that is responsible of customizing the visuals of the edges of the Pie chart called `wedgeprops`, it expect a dictionary of visual properties as keys and their corresponding values as dictionary values.    
As an example, we can define the border color around the Pie circle and around each segment by setting the visual attribute `edgecolor` to any color, let it be `black` for example.

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

we can have multiple slices values. If we only specified the slices values without passing colors values, `plt.pie()` will set color by its own.

```python
slices = [60, 40, 30, 20]
labels = ['sixty', 'forty', 'extra1', 'extra2']
```

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

The colors of the previous Pie chart was automatically generated according to the `plt.style` we have defined. To explicitly define colors, we have to pass a list of colors to an attribute of `plt.pie()` called colors, each value of the passed colors list corresponds to each slice respectively. 

```python
colors = ['blue', 'red', 'yellow', 'green']
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

We are not only limited to basic colors that are difined in string foramt, we can define color with RGB values.

```python
# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']
```

```python
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

We can emphasize a specific slice by displace it a little bit from the origin of the Pie circle. the unit of displacement is a ratio of the Pie circle radius. The attribute of plt.pie() which is responsible of this emphasis behaviour is called explode and it accept a list of emphasis values for each slice.      
For example, if we want to emphasize only Python segment, we have to set the value in the corresponding index of Python slice to emphasis value in the list of explode attribute.

```python
explode = [0, 0, 0, 0.1, 0]
```

```python
plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

There is a property of adding a shadow the Pie chart by setting the attribute `shadow` in `plt.pei()` to `True`.

```python
plt.pie(slices, labels=labels, explode=explode, 
        shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

We can control from what angle the Pie chart should start by setting the attribute `startangle` in `plt.pie()` to a desired angle. By default, matplotlib choose a random initial value for startangle value. 

```python
plt.pie(slices, labels=labels, explode=explode, startangle=90,
        shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```

We can also show the percentages of each slice by setting the attribute `autopct` to a string `'%1.1f%%'`

```python
plt.pie(slices, labels=labels, explode=explode, startangle=90,
        shadow=True, wedgeprops={'edgecolor': 'black'},
        autopct='%1.1f%%'
       )
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.show()
```
